```python
titanic.groupby('Sex').count()  # will show the count of all non null values for female and male
```

```python
titanic.groupby('Sex')[titanic.select_dtypes(include=['int','float']).columns].agg(['mean','count','max','min','std'])

# groupby the titanic dataset by 'Sex'

# since object does not support object datatype we need to extract the columns which is having
# only int/float using select_dtypes(include=['int','float'])

#using agg(), we can able to do the aggregate operations(['mean','count',..] for the selected numeric columns
```

```python
titanic.groupby('Sex').agg({'Age':['mean','count','max'],
                           'Fare':['mean','count'],
                           'Survived':'mean'})
```

```python
# grouping based on 2 columns

titanic.groupby(['Age','Sex']).count().head(27)
```

```python
# apply() for class:1--> business1 2-->business2 3-->business3

def classdiv(a):
    if a==1:
        return 'Business_1'
    elif a==2:
        return 'Business_2'
    else:
        return 'Business_3'


    
```

```python
titanic['class']=titanic['Pclass'].apply(lambda x:classdiv(x))
titanic
```

```python
def classdiv1(pclass,fare):
    if pclass==1 and fare>30:
        return 'Luxury'
    if pclass==2 and fare>20:
        return 'Premium'
    if pclass==3:
        return 'Economy'
    else:
        return 'Basic'

titanic['_CATEGORY_']=titanic.apply(lambda x:classdiv1(x['Pclass'],x['Fare']),axis=1)
titanic



# in pandas in replace(),drop() alone axis=0 means just rows and axis=1 means just columns
```
